A Poisson distribution is a discrete probability distribution that models the count of events or characteristics over a constant observation space, such as a particular length of time, area or volume.

The values must be integers that are greater than or equal to zero.

History - originally applied to the world of gambling, in 1830, by a French mathematician Simeon Denis Poisson. He developed it to indicate the low to high spread of the probable number of times that a gambler would win at a gambling game, within a large number of times that the game was played.
When to use it - the following assumptions are made:
   1.   Events occur with a constant mean rate(historical/given)
   2.   Events occur independently.

   3.   The events cannot occur simultaneously

   4.   The number of times an event happens within a specified interval is a count/ discrete/ non-continuous/ simple number

   5.   The number of trials (chances for the event to occur) is sufficiently greater than the number of times the event does actually occur i.e. the Poisson distribution is only designed to be applied to events that occur relatively rarely.
Where to use it:
Review and evaluate business insurance coverage.

- Determine the average number of losses or claims that occur each year and that are covered by the company’s business insurance.

- Then do a Poisson probability calculation to determine the maximum and minimum numbers of claims that might reasonably be filed during any one year.

- Review the cost of your insurance and the coverage it provides.

- Consider whether perhaps you’re overpaying
 that is, paying for a coverage level that you probably don’t need, given the probable maximum number of claims.

- Alternatively, you may find that you’re underinsured
 that if what the Poisson distribution shows as the probable highest number of claims actually occurred one year, your insurance coverage would be inadequate to cover the losses.
Model financial count data.

it can be used to model the number of trades that a typical investor will make in a given day, which can be 0 (often), or 1 or 2 etc

it can be used to predict the number of "shocks" to the market that will occur in a given time period, say over a decade.

THE CODE:

k = the number of time an event occurs.

lam = the expected value/historical average rate of occurence

poisson.pmf(k, lam) function - measures the probability mass function of the distribution. see poisson Distribution Formula

PMF aka discrete density function - it defines the probability distribution of a discrete random variable.

THE GRAPH:

Mode - the most probable number of events is represented by the peak of the distribution.

When ƛ is a non integer, the mode is the closest integer smaller that ƛ
When ƛ is an integer, there are two modes: ƛ and ƛ-1
When ƛ is low, the distribution is strongly right-skewed.
As ƛ increases, the probability decreases because there are much more successes expected within the same period of time
Mean, Variance and standard deviation

```python
#@title Default title text
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson
%matplotlib inline


#array of k values (Number of successes) 
k_1 = np.random.poisson(lam=200, size=1000)
k_2 = np.random.poisson(lam=499.5, size=1000)
k_3 = np.random.poisson(lam=820, size=1000)

fig, ax = plt.subplots()


#array with poisson probability mass function values

pmf_1 = poisson.pmf(k_1, 200)
pmf_2 = poisson.pmf(k_2, 499.5)
pmf_3 = poisson.pmf(k_3, 820)


#plot poisson PMF

ax.bar(k_1, pmf_1, color = "blue",label = "lam = 200")
leg = ax.legend()

plt.title('PMF of a Poisson distribution(lam 200)')
plt.xlabel('k')
plt.ylabel('probability mass function')

plt.show()


fig, ax = plt.subplots()

ax.bar(k_1, pmf_1, color = "blue",label = "lam = 200")
ax.bar(k_2, pmf_2, color = "black",label = "lam = 499.5")
ax.bar(k_3, pmf_3, color = "green",label = "lam = 820")
leg = ax.legend()

print('-------------------------------------------------------------------------')

plt.title("PMF of a Poisson Distribution")
plt.xlabel('k')
plt.ylabel('probability mass function')

plt.show()
```

Kurtosis is a measure of whether the data is heavy or lightly tailed relative to a normal distribution. positive Kurtosis means heavy tailed(many outliers) negative kurtosis means lightly tailed(or lack outliers)
